```python
import pandas as pd
import numpy as np
import os
import warnings
```

```python
# useful Sources: reddit, news
# useful columns: source, query, score, headline, summary

warnings.filterwarnings("once")

path = "../../data/SIGIR2013/"

num_articles = 1000
article_per_query = 10
num_files = int(num_article/article_per_query)

files = os.listdir(path)
number_files = len(files)
indices_files = np.random.choice(range(number_files), num_files, replace=False)

corpus = pd.DataFrame(data=None, columns=["source", "query", "score", "headline", "summary"])

for file_idx in indices_files:
    # get file randomly
    f = path + files[file_idx]

    # initilize dataframe
    df = pd.read_csv(f, header=1, encoding="utf-8", sep=",", error_bad_lines=False)
    del df["uniqueid"]
    del df["queryid"]
    del df["querydate"]
    del df["docdate"]
    df = df[np.any(np.stack((df["source"]=="reddit", df["source"]=="news"), axis=1), axis=1)]
    df.reset_index(drop=True, inplace=True)
    
    if len(df)==0:
        continue
    
    # only consider valid rows (with all information available)
    randomize_rows = []
    for row_idx in np.random.choice(range(len(df)), len(df), replace=False):
        if not np.any(pd.isnull(df.loc[row_idx])):
            randomize_rows.append(row_idx)
            
    if len(randomize_rows) < article_per_query:
        continue

    # get examples of relevant query results randomly
    for example in range(article_per_query):
        row = randomize_rows[example]
        corpus = corpus.append(df.loc[row])
```

```python
corpus.reset_index(drop=True, inplace=True)
```

```python
corpus.to_csv("eval_corpus.csv", sep=";", encoding="utf-8")
```
