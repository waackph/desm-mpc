```python
import numpy as np
import matplotlib.pyplot as pyplot
```

```python
#Now we can visualize the Documents in a 2-dim. Space, for human interpretability
y = [2.233408, 5.096941, 8.598322, 17.138467, 39.270495, 99.306344, 169.094607, 290.792387, 564.256688, 1376.994714]
x = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]

pyplot.plot(x, y)
pyplot.title("Laufzeit-Messung")
pyplot.ylabel("Laufzeit des Rankings [s]")
pyplot.xlabel("Anzahl der Dokumente")
pyplot.savefig("ranking_docs_ylinear.png")
pyplot.show()
pyplot.close()
```

```python
pyplot.plot(x, y)
pyplot.title("Laufzeit bei ansteigender Dokumentenmenge")
pyplot.ylabel("Laufzeit des Rankings [s]")
pyplot.xlabel("Anzahl der Dokumente")
pyplot.xscale("log", basex=2)
pyplot.savefig("ranking_docs_ylog.png")
pyplot.show()
pyplot.close()
```

```python
x = [[21,34,97,388,2,91,347,0,3,18],
    [21,264,69,60,0,10,110,568,36,47],
    [76,8,11,95,12,10,25,23,14,449],
    [5,1,0,147,29,6,8,2,46,126,7,86,298,26,64,152,127,25,4,214]]
y = [1,2,3,4]
label = ["whats the point", "aaron is dead", "justin timberlake", "devil may cry"]

pyplot.figure(figsize=(10,6))

pyplot.scatter(x[0], np.ones(10) * y[0])
pyplot.scatter(x[1], np.ones(10) * y[1])
pyplot.scatter(x[2], np.ones(10) * y[2])
pyplot.scatter(x[3], np.ones(20) * y[3])

pyplot.scatter(np.median(x[0]), y[0], c="orange", s=[50])
pyplot.scatter(np.median(x[1]), y[1], c="orange", s=[50])
pyplot.scatter(np.median(x[2]), y[2], c="orange", s=[50])
pyplot.scatter(np.median(x[3]), y[3], c="orange", s=[50])

pyplot.vlines(x=10, ymin=0, ymax=5, color='r')
pyplot.vlines(x=50, ymin=0, ymax=5, color='r')
pyplot.vlines(x=100, ymin=0, ymax=5, color='r')

pyplot.xlim(-50, 830)
pyplot.ylim(0, 5)

pyplot.title("Evaluation der Effektivität")
pyplot.ylabel("Suchanfrage")
pyplot.xlabel("Ranking-Position")
pyplot.yticks(y, label)
#pyplot.xscale("log", basex=2)
pyplot.savefig("efficiency_median_multilines.png")
pyplot.show()
pyplot.close()
```
